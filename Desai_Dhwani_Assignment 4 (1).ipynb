```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import ipywidgets
```

Imported matplot library,numpy,pandas and ipywidgets to carry out operations

```python
ufo1 = pd.read_csv("https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_AOUAOG_fall2021/master/week05/data/ufo-scrubbed-geocoded-time-standardized-00.csv",
                    names = ["Date", "City", "State", "Country",
                          "Shape", "Duration_seconds", "Duration",
                          "Comment", "Report_date", "Latitude", "Longitude"],
                  parse_dates = ["Date", "Report_date"])
```

```python
ufo1
```

Read the .csv file and added header for columns like city,state,country,etc. and additionaly also used the parse_date funtction to specify dates and report_dates in their specific format

```python
ufo1 = ufo1.fillna({'Country':'Unknown','State':'Unknown','Shape':'Unknown'})
```

```python
ufo1
```

Coverted all the NAN values in the column country, state and shape to Unknown values

```python
plt.plot(ufo1['Longitude'],ufo1['Latitude'],'.', color='violet')
plt.show()
```

Used the plot to illustrate the longitude and latitude with the use of a marker using a violet color

```python
plt.scatter(ufo1['Longitude'],ufo1['Latitude'],c=ufo1['Duration_seconds'])
```

Used the scatter plot to show the data of longitude and latitude

```python
nsamples = 50
downSample = np.random.choice(range(len(ufo1)-1), nsamples, replace=False)
downSample
```

Used the nsample function to get proper details of just 50 rows out of 80332 rows. This potrayed the array for 50 rows

```python
ufo2 = ufo1.loc[downSample]
```

```python
ufo2
```

Used the location function to show the location of each row 

```python
xlab = ['Date','Longitude','Latitude','Report_date','Duration_seconds','City','State','Country','Shape']
ylab = ['Date','Longitude','Latitude','Report_date','Duration_seconds']
col = ['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
            'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
            'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn']

@ipywidgets.interact(style=plt.style.available, x=xlab,y=ylab,cmap=col)
def make_plot(style,x,y,cmap):
    with plt.style.context(style):
        fig,ax = plt.subplots(1,1,figsize=(12,10))
        plt.scatter(ufo2[x],ufo2[y],c=np.log10(ufo2['Duration_seconds']), cmap = cmap)
        
        
        ax.set_ylabel(y)
        ax.set_xlabel(x)
        
    plt.show()    
```

Used the drop down menu for x and y axis to show different headers like Date, Longitude, Latitude, Report_date, etc. Made use of the interact function to potray different styles on the plot. Also, used the x_label and y_label to show the different options that have been mentioned in the dropdown of X and Y

```python
irange = ipywidgets.IntSlider(min=1, max=100, step=1) 
xn = ['Date','Report_date','Longitude','Latitude','Duration_seconds','State','Country','City','Shape']


@ipywidgets.interact(style=plt.style.available, x=xn,bins=irange)
def make_plot(style,x,bins):
    if(x=='country' or x=='shape'or x=='city' or x=='state'):
        with plt.style.context(style):
            fig, ax = plt.subplots(1,1,figsize=(10,10))
            plt.bar(ufo2[x].value_counts().index, height = ufo2[x].value_counts().values)
    else:
        with plt.style.context(style):
            fig, ax = plt.subplots(1,1,figsize=(10,10))
            plt.hist(ufo2[x], bins=bins)
        
    ax.set_xlabel(x)
plt.show()
```

To illustrate categorical values it's not really possible to make a histogram so I made use of the bar graph and if the values are numerical then it would use histogram as bins can store the numric data with good efficiency.
