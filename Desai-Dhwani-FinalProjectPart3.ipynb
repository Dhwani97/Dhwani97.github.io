```python
analysis_final = pd.pivot_table(Crime, index = 'Highest NIBRS/UCR Offense Description', columns = 'Council District', values = 'GO Census Tract', aggfunc = 'mean', fill_value=0)
```

Used the pivot table to create the Heat Map.Highest NIBRS/UCR Offense Description is used to specify the various offfences that take place in Austin which are agg Assault, Auto theft, Burglary, Murder, Rape, Robbery and Theft.The column I used was Council District as it specifies the council district of specific states and the value which I used was the population count of those specific states. The aggregate function which I used was Mean.

```python
analysis_final
```

Here is the pivot table.

```python
x_scs = bqplot.OrdinalScale()
y_scs = bqplot.LinearScale()

x_axs = bqplot.Axis(label='Offense Description', scale=x_scs)
y_axs = bqplot.Axis(label='Count', scale=y_scs, 
                   orientation = 'vertical')

bar_plot = bqplot.pyplot.bar(x = [], 
                         y = [],
                         scales={'x':x_scs, 'y':y_scs}, colors='orange')

fig_dur = bqplot.Figure(marks=[bar_plot], axes=[x_axs,y_axs], fig_margin={'top':20, 'bottom':80, 'left':40, 'right':80})
```

Used the bar plot to specify the offense Description and its count. I also changed the color to orange and set the margins for top, bottom, left and right as 20,80,40 and 80 respectively.

# Different Offenses in specific district in Austin

```python
#Scales
col_sc = bqplot.ColorScale(scheme='Red', min=np.nanmin(analysis_final.values), max=np.nanmax(analysis_final.values))
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale() 

#Axis
col_ax = bqplot.ColorAxis(scale=col_sc, orientation='vertical', side='right')
x_ax = bqplot.Axis(scale=x_sc, label = 'Council District')
y_ax = bqplot.Axis(scale=y_sc, orientation='vertical', label = 'Offense Description')

#Marks
heat_map = bqplot.GridHeatMap(color=analysis_final.values, 
                             row = analysis_final.index, column = analysis_final.columns, 
                             scales={'color':col_sc, 'row':y_sc, 'column':x_sc}, 
                             interactions={'click':'select'}, 
                             anchor_style={'fill':'black'})

def barplot_fig2(x, y):
    df2 = Crime[(Crime['Council District'] == y) & (Crime['Highest NIBRS/UCR Offense Description'] == x)]
    df2 = df2[df2['GO Highest Offense Desc'] != 0]
    df2 = df2['GO Highest Offense Desc'].value_counts()
    return df2

mySelectedLabel = ipywidgets.Label()
def on_selected(change):
    if len(change['owner'].selected) == 1: 
        i,j = change['owner'].selected[0]
        v = analysis_final.values[i,j] 
        mySelectedLabel.value = 'Mean Census Tract for: ' + str(analysis_final.index[i]) + ', and Council District No.' + str(analysis_final.columns[j]) + ' is ' + str(v)
        analysis_final2 = barplot_fig2(analysis_final.index[i], analysis_final.columns[j])
        if len(analysis_final.index[i]) > 0:
            bar_plot.x = analysis_final2.index
            bar_plot.y = analysis_final2.values
        else:
            bar_plot.x = []
            bar_plot.y = analysis_final2.values
heat_map.observe(on_selected, 'selected')

# 6. figure!
fig = bqplot.Figure(marks=[heat_map], axes=[col_ax, x_ax, y_ax], fig_margin={'top':30, 'bottom':60, 'left':80, 'right':100})

myDashboard = ipywidgets.VBox([mySelectedLabel, fig, fig_dur])
myDashboard
```

This dashboard mainly illustrates the different types of offenses that take place and in which specific states.I further segregated them into the exact type of offense that was seen in that perticular state. For eg, if it was a burglary, it would then be segregated into if it was burglary of residence or burglary on non-residence, Burglary of non residence sheds or burglary of residence sexual nature. This would give the user great clarity on the different types of crimes that took place. For this, used the interactivity between the heat map and the bar plot so that on one click I could recieve the segregated offenses for better clarity. Also,used the label above the HeatMap to show Mean census tract for the specific offense the exact council district.
